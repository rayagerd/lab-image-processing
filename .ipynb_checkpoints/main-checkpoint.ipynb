# Channels and color spaces

### **Exercise: Image Creation and Color Manipulation**

*   Create a 100 x 100 image for each of the below visualization



*   Visualize the created images in a 1 x 3 subplot using matplotlib.


```python
# solution
img1 = np.zeros((100, 100))
img2 = np.zeros((100, 100))
img3 = np.zeros((100, 100))

# Right half white
img1[:, 50:] = 1

# Bottom half white
img2[50:, :] = 1

# Top-left square white
img3[:50, :50] = 1
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 4))

axes[0].imshow(img1, cmap='gray')
axes[0].set_title("Right half white")

axes[1].imshow(img2, cmap='gray')
axes[1].set_title("Bottom half white")

axes[2].imshow(img3, cmap='gray')
axes[2].set_title("Top-left square white")

for ax in axes:
    ax.set_xlabel("X axis")
    ax.set_ylabel("Y axis")

plt.tight_layout()
plt.show()
```

*   Use the above three images to create the following image


*Hint: Remember channels and color spaces*

```python
# solution
color_img = np.zeros((100, 100, 3))

# Red channel
color_img[:, :, 0] = img1

# Green channel
color_img[:, :, 1] = img2

# Blue channel
color_img[:, :, 2] = img3

plt.imshow(color_img)
plt.title("Combined RGB image")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

### **Exercise: Color Manipulation**

*   Read the image 'sillas.jpg' from the images folder



*   Extract individual channels and plot them using matplotlib subplot.



```python
# solution
from PIL import Image
from numpy import asarray
image = Image.open('images/sillas.jpg')
img = asarray(image)
```

*   The color **red** looks too bright for the eyes. Isn't it?? Lets change the color and see how it appears.
    *    Create a new image where everything that is **'red' is changed to 'blue'**.
*   Visualize the original image and the created image using matplotlib subplot.

```python
# solution
blue_img = img.copy()

# Extract channels
red = blue_img[:, :, 0]
green = blue_img[:, :, 1]
blue = blue_img[:, :, 2]

# Detect red pixels
mask = (red > 150) & (green < 100) & (blue < 100)

# Change red to blue
blue_img[mask] = [0, 0, 255]

# Plot original and modified image
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].imshow(img)
axes[0].set_title("Original Image")

axes[1].imshow(blue_img)
axes[1].set_title("Red changed to Blue")

plt.tight_layout()
plt.show()
```

# Image Manipulation

### **Exercise: Image Operators**

*   You can find images 'model.png' and 'coat.png' in the images folder (First two images of the below visualization). Your task is to create an image from the given two images such a way that the model is wearing the coat (Third image in the visualization).
*   You can also find different textures in the images folder. Your task is to change the coat texture to any one of the given textures.
*   Visualize the images similar to the given visualization.

*Hint: Think masks!!!*



```python
# solution

coat = io.imread("images/coat.png")
model = io.imread("images/model.png")

# copy model so we don't destroy original
model_with_coat = model.copy()

# create mask: where coat is not black
mask = np.any(coat[:, :, :3] > 20, axis=2)

# paste coat pixels onto model
model_with_coat[mask] = coat[mask]

# visualize
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(coat)
axes[0].set_title("Coat")

axes[1].imshow(model)
axes[1].set_title("Model")

axes[2].imshow(model_with_coat)
axes[2].set_title("Model with Coat")

plt.tight_layout()
plt.show()
```

```python
from skimage.transform import resize

texture = io.imread("images/texture3.jpg")

# resize texture to same height/width as coat
texture_resized = resize(
    texture,
    coat.shape,
    preserve_range=True
).astype(coat.dtype)

textured_model = model.copy()

# Apply texture only where coat exists
textured_model[mask] = texture_resized[mask]

plt.imshow(textured_model)
plt.title("Model with Textured Coat")
plt.show()
```

# Contrast Enhancement

### **Exercise: Histogram Computation**

*   Read the **'astronaut' image** from data module.
*   Convert the image to grayscale.
*   Compute the **histogram of the image.** *Hint: histogram function is available in skimage.exposure package*
*   Plot the histogram using matplotlib plot.




```python
# solution
from skimage.color import rgb2gray

# Read astronaut image
img = data.astronaut()

# Convert to grayscale
gray_img = rgb2gray(img)

# Compute histogram
hist, bins = exposure.histogram(gray_img)

# Plot grayscale image
plt.figure(figsize=(6,6))
plt.imshow(gray_img, cmap='gray')
plt.title("Grayscale Astronaut")
plt.axis('off')
plt.show()

# Plot histogram
plt.figure(figsize=(8,5))
plt.plot(bins, hist)

plt.title("Histogram of Grayscale Image")
plt.xlabel("Pixel intensity")
plt.ylabel("Frequency")

plt.show()
```

*   Change the bin count to 8 and compute the histogram of the image and plot the computed histogram using matplotlib plot.

```python
# solution
hist, bins = exposure.histogram(gray_img, nbins=8)

plt.figure(figsize=(8,5))
plt.plot(bins, hist)

plt.title("Histogram with 8 bins")
plt.xlabel("Pixel intensity")
plt.ylabel("Frequency")

plt.show()
```



*   What happens when you change the bin count? Does your inference change based on the bin count? If yes, then how do you define the correct bin count.

When the bin count changes, the histogram representation changes as well.
Yes, the inference can change depending on the number of bins because the image intensity values are grouped differently.
The correct bin count depends on the level of detail needed and the type of image being analyzed.


* What happens when the bin count is very low and what happens when it is very high?

With a low bin count:
- The histogram becomes smoother
- Small intensity differences are grouped together
- Some image details may be lost

With a high bin count:
- The histogram becomes more detailed
- Intensity distribution is more precise
- The histogram can become noisy and harder to interpret
